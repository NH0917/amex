```python
def xgb_amex_metric(y_pred, y_true):
    y_true = y_true.get_label()
    return 'amex_metric', amex_metric_xgb(y_true, y_pred)

def amex_metric_xgb(y_true, y_pred):
    labels = np.transpose(np.array([y_true, y_pred]))
    labels = labels[labels[:, 1].argsort()[::-1]]
    weights = np.where(labels[:,0]==0, 20, 1)
    cut_vals = labels[np.cumsum(weights) <= int(0.04 * np.sum(weights))]
    top_four = np.sum(cut_vals[:,0]) / np.sum(labels[:,0])
    gini = [0,0]
    for i in [1,0]:
        labels = np.transpose(np.array([y_true, y_pred]))
        labels = labels[labels[:, i].argsort()[::-1]]
        weight = np.where(labels[:,0]==0, 20, 1)
        weight_random = np.cumsum(weight / np.sum(weight))
        total_pos = np.sum(labels[:, 0] *  weight)
        cum_pos_found = np.cumsum(labels[:, 0] * weight)
        lorentz = cum_pos_found / total_pos
        gini[i] = np.sum((lorentz - weight_random) * weight)
    return 0.5 * (gini[1]/gini[0] + top_four)


def xgb_model(train_df,val_df,output_path,n_fold):
  feature_col = ["prediction_x","prediction_y","oof_pred"]
  dtrain = xgb.DMatrix(train_df[feature_col],label=train_df["target"])
  dvalid = xgb.DMatrix(val_df[feature_col],label=val_df["target"])

  params = {
      "objective":"binary:logistic",
           }

  model = xgb.train(params,dtrain,num_boost_round=10000,early_stopping_rounds=100,evals=[(dtrain,"train"),(dvalid,"valid")],feval=xgb_amex_metric,maximize=True)

  train_predict_df = train_df[["customer_ID","target"]]
  val_predict_df = val_df[["customer_ID","target"]]
  
  train_predict_df.loc[:,"xgb_predict"] = model.predict(dtrain)
  val_predict_df.loc[:,"xgb_predict"] = model.predict(dvalid)

  model.save_model(os.path.join(output_path,f"xgb_model{n_fold}.pkl"))

  return train_predict_df,val_predict_df
```

```python
def amex_metric(y_true, y_pred):
    labels = np.transpose(np.array([y_true, y_pred]))
    labels = labels[labels[:, 1].argsort()[::-1]]
    weights = np.where(labels[:,0]==0, 20, 1)
    cut_vals = labels[np.cumsum(weights) <= int(0.04012353 * np.sum(weights))]
    top_four = np.sum(cut_vals[:,0]) / np.sum(labels[:,0])
    gini = [0,0]
    for i in [1,0]:
        labels = np.transpose(np.array([y_true, y_pred]))
        labels = labels[labels[:, i].argsort()[::-1]]
        weight = np.where(labels[:,0]==0, 20, 1)
        weight_random = np.cumsum(weight / np.sum(weight))
        total_pos = np.sum(labels[:, 0] *  weight)
        cum_pos_found = np.cumsum(labels[:, 0] * weight)
        lorentz = cum_pos_found / total_pos
        gini[i] = np.sum((lorentz - weight_random) * weight)
    return 0.5 * (gini[1]/gini[0] + top_four)

def amex_metric_np(preds, target):
    indices = np.argsort(preds)[::-1]
    preds, target = preds[indices], target[indices]
    weight = 20.0 - target * 19.0
    cum_norm_weight = (weight / weight.sum()).cumsum()
    four_pct_mask = cum_norm_weight <= 0.04
    d = np.sum(target[four_pct_mask]) / np.sum(target)
    weighted_target = target * weight
    lorentz = (weighted_target / weighted_target.sum()).cumsum()
    gini = ((lorentz - cum_norm_weight) * weight).sum()
    n_pos = np.sum(target)
    n_neg = target.shape[0] - n_pos
    gini_max = 10 * n_neg * (n_pos + 20 * n_neg - 19) / (n_pos + 20 * n_neg)
    g = gini / gini_max
    return 0.5 * (g + d)

def lgb_amex_metric(y_pred, y_true):
    y_true = y_true.get_label()
    return 'amex_metric', amex_metric(y_true, y_pred), True


def lgbm_model(train_df,val_df,output_path,n_fold):
    feature_col = ["prediction_x","prediction_y","oof_pred"]
    dtrain = lgb.Dataset(train_df[feature_col],train_df["target"])
    dvalid = lgb.Dataset(val_df[feature_col],val_df["target"])

    params = {
        'objective': 'binary',
        'metric': "binary_logloss",
        
        }

    model = lgb.train(
            params = params,
            train_set = dtrain,
            num_boost_round = 10500,
            valid_sets = [dtrain, dvalid],
            early_stopping_rounds = 1000,
            verbose_eval = 500,
            feval = lgb_amex_metric
            )
    
    train_predict_df = train_df[["customer_ID","target"]]
    val_predict_df = val_df[["customer_ID","target"]]


    train_predict_df.loc[:,"lgbm_predict"] = model.predict(train_df[feature_col])
    val_predict_df.loc[:,"lgbm_predict"] = model.predict(val_df[feature_col])

    joblib.dump(model,os.path.join(output_path,f"lgbm_fold{n_fold}.pkl"))

    return train_predict_df,val_predict_df
```

```python
def svm_model(train_df,val_df,output_path,n_fold):
  feature_col = ["prediction_x","prediction_y","oof_pred"]

  model = SVC(cache_size=1024*10,probability=True)
  model.fit(train_df[feature_col],train_df["target"])
  linear_path = f"/content/drive/MyDrive/kaggle/amex/ensemble/cal_score/cal_ensemble/linear_{n_fold}.pkl"
  linear_model = pickle.load(linear_path)

  train_predict_df = train_df[["customer_ID","target"]]
  val_predict_df = val_df[["customer_ID","target"]]

  train_predict_df.loc[:,"svm_predict"] = model.predict_proba(train_df[feature_col].values)[:,-1]
  val_predict_df.loc[:,"svm_predict"] = model.predict_proba(val_df[feature_col].values)[:,-1]

  save_name = os.path.join(output_path,f"svm_{n_fold}.pkl")

  pickle.dump(model,open(save_name,"wb"))

  return train_predict_df,val_predict_df
```

```python
def linear_model(train_df,val_df,output_path,n_fold):
    feature_col = ["prediction_x","prediction_y","oof_pred"]

    
    model = Ridge()
    model = model.fit(train_df[feature_col],train_df["target"])

    train_predict_df = train_df[["customer_ID","target"]]
    val_predict_df = val_df[["customer_ID","target"]]

    train_predict_df.loc[:,"linear_predict"] = model.predict(train_df[feature_col].values)
    val_predict_df.loc[:,"linear_predict"] = model.predict(val_df[feature_col].values)

    save_name = os.path.join(output_path,f"linear_{n_fold}.pkl")

    pickle.dump(model,open(save_name,"wb"))

    return train_predict_df,val_predict_df
```

```python
def load_predict_df():
  model1_predict_path = "/content/drive/MyDrive/kaggle/amex/ensemble/model1/output/test_lgbm_baseline_5fold_seed42.csv"
  model3_predict_path = "/content/drive/MyDrive/kaggle/amex/ensemble/model3/output/predictions_test.csv"
  model4_predict_path = "/content/drive/MyDrive/kaggle/amex/ensemble/model4/output/submission_xgb_v1.csv"

  model1_df = pd.read_csv(model1_predict_path)
  model3_df = pd.read_csv(model3_predict_path)
  model4_df = pd.read_csv(model4_predict_path)

  model_df = model1_df.merge(model3_df,on="customer_ID").merge(model4_df,on="customer_ID")
  pred_df = model_df[["customer_ID"]]
  sub_df = model_df[["customer_ID"]]

  return model_df,pred_df,sub_df

def predict_lgbm(model_df,pred_df,n_fold):

  lgbm_path = f"/content/drive/MyDrive/kaggle/amex/ensemble/cal_score/cal_ensemble/lgbm_fold{n_fold}.pkl"
  lgbm = joblib.load(lgbm_path)

  pred_df.loc[:,f"lgbm_pred{n_fold}"] = lgbm.predict(model_df.drop(columns="customer_ID"))

  return pred_df

def predict_xgb(model_df,pred_df,n_fold):

  xgb_path = f"/content/drive/MyDrive/kaggle/amex/ensemble/cal_score/cal_ensemble/xgb_model{n_fold}.pkl"
  xgb_model = xgb.Booster()
  xgb_model.load_model(xgb_path)

  pred_df.loc[:,"xgb_pred"] = xgb.predict(model_df.drop(columns="customer_ID"))

  return pred_df

def predict_svm(model_df,pred_df,n_fold):

  svm_path = f"/content/drive/MyDrive/kaggle/amex/ensemble/cal_score/cal_ensemble/svm_{n_fold}.pkl"
  svm_model = pickle.load(svm_path)

  pred_df.loc[:,"svm_pred"] = svm_model.pred(model_df.drop(columns="customer_ID"))

  return pred_df

def predict_linear(model_df,pred_df,n_fold):

  linear_path = f"/content/drive/MyDrive/kaggle/amex/ensemble/cal_score/cal_ensemble/linear_{n_fold}.pkl"
  linear_model = pickle.load(linear_path)

  pred_df.loc[:,"linear_pred"] = linear_model.pred(model_df.drop(columns="customer_ID"))

  return pred_df

def predict_stack2(sub_df,pred_df,n_fold):

  stack2_path = f"/content/drive/MyDrive/kaggle/amex/ensemble/cal_score/cal_ensemble/lgbm_stacking_model_fold{n_fold}.pkl"
  stack2 = joblib.load(stack2_path)

  sub_df.loc[:,f"sub_fold{n_fold}"] = stack2.predict(pred_df.drop(columns="customer_ID"))

  return sub_df

```
